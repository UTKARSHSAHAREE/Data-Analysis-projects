## Normal Distribution

```python
Samples = np.random.normal(loc =1,scale=2,size=500)
```

```python
plt.figure(figsize=(7,6))
sns.histplot(data=Samples,color=color)
plt.title('Normal Distribution')
plt.show()
```

#### loc (Location): This represents the mean (average) of the distribution, which determines where the center of the "bell curve" is located on the x-axis.
#### scale (Scale): This represents the standard deviation, which determines the "spread" or width of the distribution. A larger scale results in a flatter, wider curve.

## Outliers in Normal Distribution

```python
plt.figure(figsize=(8,7))
sns.boxplot(data=Samples,color=color,orient='h') # orient(orientation) = 'h'horizontal
plt.title('Outliers in Normal Distribution')
plt.show()
```

### Observation
#### It is way to visualize how your data is spread out.
#### The Box: This dark red box represents the middle 50% of your data. In your plot, it spans roughly from 0 to 2.5. This is called the Interquartile Range (IQR).
#### The Center Line: The vertical line inside the box is the median (the exact middle value). It is sitting right around 1, which matches the loc=1 parameter you used in your code.
#### The Whiskers: The horizontal lines sticking out of the box are called whiskers. They show the typical range for most of your data.
#### Anything within these whiskers is considered "normal" or expected variation for this group
#### The Dots: Those individual diamonds on the far left and far right are the outliers.
#### In a normal distribution, these are rare values that landed very far from the center (mean).

### Why are they there? Even though your distribution is "normal," random chance will occasionally produce a very high or very low number that doesn't fit with the rest of the pack.

### Outlier De-emphasis:The transformation pulls extreme values toward the mean, reducing their disproportionate influence on your statistical

### Variance Stabilization: It helps meet the assumption of "constant variance" (homoscedasticity) required for linear modeling by reducing the spread of data across different scales.

### Linearization: Log transformations can convert non-linear, exponential relationships into linear ones, which is a requirement for many machine learning and regression models.

### Symmetry and Normality: An exponential distribution is highly right-skewed, meaning it has a long tail of large values. Taking the log "compresses" these large values and "stretches" smaller ones, often resulting in a more symmetric, bell-shaped distribution similar to a normal distribution.

# Function for Outlier Tag


```python
def feature_outlier_tagger(df,feat):
    # calculate 25th percentile and 75th percentile
    quartile25=np.percentile(df[feat],25)
    quartile75=np.percentile(df[feat],75)

    # calculate IQR interquartile Range
    IQR = quartile75-quartile25
    
    # create a new column in  DataFrame to store outlier tags for feature **imp
    df[f'{feat}_outlier'] = df[feat].apply(outlier_tag, 
                                                lower_bound = (quartile25-IQR*1.5),
                                                upper_bound = (quartile75+IQR*1.5))
    # count the number of Unique outlier tags 
    num_cat = len(df[f'{feat}_outlier'].value_counts())

    # plot a histogram of the feature,colored by the outlier tags
    sns.histplot(data=df,x=feat,hue=f'{feat}_outlier',palette=[color,color2][:num_cat])

    # set the title for the plot
    plt.title(feat,fontsize=20,fontweight='bold')

    # Disply the plot
    plt.show()
    return df
```

## def binner

```python
def binner(df,feat,bins=3):
    # calculate the Edges for binning based on the specified number of bins
    #Imagine you have a huge bag of mixed-size pebbles, and you want to organize them into a few buckets (like "small," "medium," and "large") so they are easier to handle.
    
    edges = np.linspace(df[feat].min(),df[feat].max(),bins+1)
    # define labels for bins (you can customize this labels)
    #The code looks at the smallest number and the largest number in your column. It then draws "lines in the sand" to create equal-sized gaps.
    # Example: If your numbers go from 0 to 30 and you want 3 bins, the "edges" will be 0, 10, 20, and 30.
    
    labels = ['a','b','c','d','e']
    # create a new columns in DataFrame to store the bin labels for the feature
    # This is just a list of names for your buckets. Instead of calling them "Bucket 1," the code is prepared to call them "a," "b," and so on.
    
    
    df[f'{feat}_bin'] = pd.cut(df[feat],bins=edges,labels=[labels[i] for i in range (bins)])
    # return the DataFrame with added bins labels
    # This is where the magic happens. The code goes through every row in your column and asks: "Which two edges does this number fall between?"
    # If your number is 5, it goes into the 'a' bucket.
    # If your number is 25, it goes into the 'c' bucket
    # It then creates a brand new column in your table to store these labels
    
    return df
    # Finally, the tool gives you back your original table, but now it has that extra column with all your numbers neatly categorized.
```

# Understanding the "Speeding" Data Distribution

This chart is a **Kernel Density Estimate (KDE) plot**. Think of it as a "smoothed-out" version of a bar chart (histogram) that shows us where data points are most common.

### 1. How to Read the Axes
*   **Horizontal Axis (`speeding`):** This shows the speed values. As you move from left to right (0 to 12), the speeding values are getting higher.
*   **Vertical Axis (Density):** This tells us how "crowded" the data is at a certain speed. 
    *   **High Peaks** = Many data points share this speed (very common).
    *   **Low Valleys** = Very few data points have this speed (rare).

### 2. Comparing the Groups (a, b, c, d)
Each colored line represents a different group of data. Their "shapes" tell different stories:

*   **Group 'b' (Orange):** This group is the most "consistent." It has a very high, narrow peak around **4**. This means most people in this group were speeding at exactly that level.
*   **Group 'c' (Green):** This is more spread out. Its "hill" is lower and wider, peaking around **7**. This means there is more variety in how fast people in this group were going.
*   **Group 'a' (Blue):** This group is unique because it has **two small peaks** (around 2). This suggests there might be two distinct "sub-groups" within group 'a'.
*   **Group 'd' (Red):** This is the most "flat" and "unpredictable" group. The data is spread thin across almost the entire range, meaning people's speeds were all over the place.

### Summary Key

| Feature | What it means in plain English |
| :--- | :--- |
| **Peak Height** | How common that specific speed is. |
| **Peak Width** | How much variety or "spread" there is in the group. |
| **Total Area** | The area under each curve always adds up to 1 (100% of that group's data). |
